## Basic exploration of the data
- We can create a vector (a series of values) listing how many species from each taxa have been recorded in Edinburgh
- We will filter out the data for each taxon group and then count the unique species within it
- Here we are only hinting to all the cool stuff we can do with pandas - we will learn more in later tutorials.

```python
# Filtering the data for each taxon group
beetle = edidiv[edidiv['taxonGroup'] == "Beetle"]
bird = edidiv[edidiv['taxonGroup'] == "Bird"]
butterfly = edidiv[edidiv['taxonGroup'] == "Butterfly"]
dragonfly = edidiv[edidiv['taxonGroup'] == "Dragonfly"]
flowering_plants = edidiv[edidiv['taxonGroup'] == "Flowering.Plants"]
fungus = edidiv[edidiv['taxonGroup'] == "Fungus"]
hymenopteran = edidiv[edidiv['taxonGroup'] == "Hymenopteran"]
lichen = edidiv[edidiv['taxonGroup'] == "Lichen"]
liverwort = edidiv[edidiv['taxonGroup'] == "Liverwort"]
mammal = edidiv[edidiv['taxonGroup'] == "Mammal"]
mollusc = edidiv[edidiv['taxonGroup'] == "Mollusc"]

# To find out the number of different species in each taxa, we will use the nunique() function 
# which counts unique values (equivalent to length(unique()))
a = beetle['taxonName'].nunique()
b = bird['taxonName'].nunique()
c = butterfly['taxonName'].nunique()
d = dragonfly['taxonName'].nunique()
e = flowering_plants['taxonName'].nunique()
f = fungus['taxonName'].nunique()
g = hymenopteran['taxonName'].nunique()
h = lichen['taxonName'].nunique()
i = liverwort['taxonName'].nunique()
j = mammal['taxonName'].nunique()
k = mollusc['taxonName'].nunique()

# We can now combine all those objects into a pandas Series (equivalent to a named vector in R)
biodiv = pd.Series(
    [a, b, c, d, e, f, g, h, i, j, k], 
    index=["Beetle", "Bird", "Butterfly", "Dragonfly", "Fl.Plants", 
           "Fungus", "Hymenopteran", "Lichen", "Liverwort", "Mammal", "Mollusc"])

# We have all the values now, so we can visualise them quickly with a bar plot
biodiv.plot(kind='bar')
plt.show()
```

## Handling bigger sets of data
This was a Series of values, each with a label, which is suitable when dealing with just one set of values. In most cases you will have more variables and categories. For that we will use data frames - we can save a data frame as a csv file to use again later

```python
# Declaring the taxa categories as a list
taxa = ["Beetle", "Bird", "Butterfly", "Dragonfly", "Fl.Plants", 
        "Fungus", "Hymenopteran", "Lichen", "Liverwort", "Mammal", "Mollusc"]
richness = [a, b, c, d, e, f, g, h, i, j, k]

# Pandas handles categorical data intrinsically or treats them as objects (strings). 
# We can explicitly set it to a category type (similar to factor in R)
taxa_f = pd.Categorical(taxa)

# Creating a DataFrame from the taxa and richness lists
biodata = pd.DataFrame({'taxa_f': taxa_f, 'richness': richness})
```

```python
# Displaying the DataFrame
biodata
```

### Saving the data frame as a csv file

```python
# Exporting the data to a CSV file
biodata.to_csv("biodata.csv", index=False) # index=False prevents writing row numbers to the CSV
```

### Plotting from the data frame instead of Series
If we want to make the same barplot using the data frame, not the Series, we need to slightly change the code. We need to tell matplotlib exactly what we want it to plot, in our case the richness on the y-axis, and taxa on x

```python
# Creating the plot from the data frame
plt.figure(figsize=(9.5, 5))
plt.bar(biodata['taxa_f'], biodata['richness'], color='gray')
plt.xlabel("Taxa", fontsize=15)
plt.ylabel("Number of species", fontsize=15)
plt.ylim(0, 600)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.tight_layout()
```

## Quick note on the "pythonic" way
In the tutorial, we create 11 separate variables (a through k) and 11 separate filtered datasets. This is done on purpose in Workshop 1 to explicitly teach how filtering and assignment work!

However, as a side note for your learning, you would typically do this entire sequence in just two lines of code using groupby.

```python
# The fast, "Pythonic" way to get species richness by taxa:
biodata_fast = edidiv.groupby('taxonGroup')['taxonName'].nunique().reset_index()
biodata_fast.columns = ['taxa', 'richness']
```
